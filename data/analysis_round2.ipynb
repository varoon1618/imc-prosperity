```python
import pandas as pd
import statsmodels.api as sm

def calculate_rolling_beta(crossPrices, bench, window=20):
    # Step 1: Compute daily returns
    stock_returns = crossPrices.pct_change()
    bench_returns = bench.pct_change()

    # Step 2: Align and clean data
    data = pd.DataFrame({
        'stock': stock_returns,
        'bench': bench_returns
    }).dropna()

    # Step 3: Rolling regression to get beta
    betas = []
    for i in range(window, len(data)):
        y = data['stock'].iloc[i - window:i]
        x = data['bench'].iloc[i - window:i]
        x = sm.add_constant(x)
        model = sm.OLS(y, x).fit()
        beta = model.params['bench']
        betas.append(beta)

    # Pad with NaNs at start to keep index aligned
    beta_series = [None] * window + betas
    data['beta'] = beta_series

    return data[['stock', 'bench', 'beta']]
```

```python
bench = djembe['mid_price'] + jam['mid_price'] + cross['mid_price']
crossPrices = cross['mid_price']
jamPrices =  jam['mid_price']
djembePrices = djembe['mid_price']
cross_beta_df = calculate_rolling_beta(crossPrices, bench, window=20)
jam_beta_df = calculate_rolling_beta(jamPrices, bench, window=20)
djembe_beta_df = calculate_rolling_beta(djembePrices, bench, window=20)

# Optional: plot the beta values
import matplotlib.pyplot as plt
jam_beta_df['beta'].plot(title='Rolling Beta (Window=20)', figsize=(10,5))
plt.show()
djembe_beta_df['beta'].plot(title='Rolling Beta (Window=20)', figsize=(10,5))
plt.show()
cross_beta_df['beta'].plot(title='Rolling Beta (Window=20)', figsize=(10,5))
plt.axhline(1, color='red', linestyle='--')
plt.ylabel('Beta')
plt.show()
```
